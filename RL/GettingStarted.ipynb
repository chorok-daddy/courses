### 샘플코드 돌려보기 (Random Cartpole)
* Cartpole에 Random한 제어 입력을 하는 형태로 시뮬레이션을 수행
* 제대로 된 제어가 되지 않으므로 1~2회만에 pole이 쓰러지고 episode가 끝나게 됨


```python
def run_random_agent(env_name="CartPole-v1", max_episodes=2, max_steps=200):
    env = gym.make(env_name)
    for episode in range(max_episodes):
        state = env.reset()
        total_reward = 0
        for step in range(max_steps):
            action = env.action_space.sample()  # 랜덤 행동
            next_state, reward, done, info = env.step(action)
            total_reward += reward
            state = next_state
            if done:
                print(f"Episode {episode+1} finished after {step+1} steps, total reward: {total_reward}")
                break
    env.close()

def tensor_example():
    x = torch.randn((3,4))
    print("PyTorch 텐서 예제:")
    print(x)
    print("텐서 크기:", x.shape)

if __name__ == "__main__":
    run_random_agent()
    tensor_example()
```
