## âš™ï¸ ëª¨ë“ˆ 3: ë‚˜ë§Œì˜ ì½”ë“œ ë¶€í’ˆ, í•¨ìˆ˜(def)

````python
# ë‘ ìˆ«ìë¥¼ ë”í•˜ëŠ” ê°„ë‹¨í•œ í•¨ìˆ˜
def add_numbers(x, y):
    """
    ì´ í•¨ìˆ˜ëŠ” ë‘ ê°œì˜ ìˆ«ì xì™€ yë¥¼ ì…ë ¥ë°›ì•„,
    ê·¸ í•©ì„ ê³„ì‚°í•˜ì—¬ ê²°ê³¼ê°’ìœ¼ë¡œ ëŒë ¤ì¤ë‹ˆë‹¤.
    """
    result = x + y
    return result

# í•¨ìˆ˜ í˜¸ì¶œ (ì‚¬ìš©)
sum_result = add_numbers(5, 3)
print(f"5ì™€ 3ì˜ í•©ì€? {sum_result}")

another_sum = add_numbers(100, 200)
print(f"100ê³¼ 200ì˜ í•©ì€? {another_sum}")
````

````python
# ëª¨ë“ˆ 2ì—ì„œ í–ˆë˜ 'í‰ì  ê¸°ë°˜ ì˜í™” ì¶”ì²œ' ë¡œì§ì„ í•¨ìˆ˜ë¡œ ë§Œë“¤ì–´ ë´…ì‹œë‹¤.

def recommend_movies(dataset, min_rating):
    """
    ì˜í™” ë°ì´í„°ì…‹ê³¼ ìµœì†Œ í‰ì ì„ ì…ë ¥ë°›ì•„,
    ì¡°ê±´ì— ë§ëŠ” ì˜í™” ì œëª© ë¦¬ìŠ¤íŠ¸ë¥¼ ê²°ê³¼ë¡œ ëŒë ¤ì¤ë‹ˆë‹¤.
    """
    recommended_list = [] # ì¶”ì²œ ëª©ë¡ì„ ë‹´ì„ ë¹ˆ ë¦¬ìŠ¤íŠ¸
    for movie in dataset:
        if movie['rating'] >= min_rating:
            recommended_list.append(movie['title']) # ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    return recommended_list


# í•¨ìˆ˜ë¥¼ ì´ìš©í•´ í‰ì  9.0 ì´ìƒ ì˜í™” ì¶”ì²œë°›ê¸°
my_best_movies = recommend_movies(my_movie_dataset, 9.0)
print(f"í‰ì  9.0 ì´ìƒ ì¶”ì²œ ì˜í™”: {my_best_movies}")

# ê°™ì€ í•¨ìˆ˜ë¥¼ ì¬ì‚¬ìš©í•´ì„œ í‰ì  9.1 ì´ìƒ ì˜í™” ì¶”ì²œë°›ê¸°
my_good_movies = recommend_movies(my_movie_dataset, 9.1)
print(f"í‰ì  9.1 ì´ìƒ ì¶”ì²œ ì˜í™”: {my_good_movies}")
````
