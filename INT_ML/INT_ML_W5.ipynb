- **íŒŒì´ì¬ ë¦¬ìŠ¤íŠ¸ì˜ ë¬¸ì œì **
  * ë¦¬ìŠ¤íŠ¸ì— ë‹´ê¸´ ëª¨ë“  ìˆ«ìì— í•œ ë²ˆì— 5ë¥¼ ë”í•˜ëŠ” ê°„ë‹¨í•œ ì—°ì‚°ì´ ë¶ˆê°€ëŠ¥

    * ë¦¬ìŠ¤íŠ¸ì™€ ìˆ«ìëŠ” ë”í•  ìˆ˜ ì—†ìŒ
- **í•´ê²°ì±…**
  * `for` ë°˜ë³µë¬¸ì„ ì‚¬ìš©í•´ì„œ ê° ìˆ«ìë¥¼ í•˜ë‚˜ì”© êº¼ë‚´ 5ë¥¼ ë”í•˜ê³ ,
  * ìƒˆë¡œìš´ ë¦¬ìŠ¤íŠ¸ì— ë‹¤ì‹œ ë‹´ì•„ì•¼ í•©ë‹ˆë‹¤.

* ë°ì´í„°ê°€ ë§ì•„ì§ˆìˆ˜ë¡ ë§¤ìš° ë²ˆê±°ë¡­ê³  ëŠë ¤ì§‘ë‹ˆë‹¤!!

````python
# íŒŒì´ì¬ ë¦¬ìŠ¤íŠ¸ë¡œ ì ìˆ˜ ê´€ë¦¬
scores_list = [80, 95, 72, 100, 88]

# ëª¨ë“  ì ìˆ˜ë¥¼ 5ì ì”© ì˜¬ë¦¬ê¸° ìœ„í•œ ê³¼ì •
new_scores_list = [] # ìƒˆë¡œìš´ ì ìˆ˜ë¥¼ ë‹´ì„ ë¹ˆ ë¦¬ìŠ¤íŠ¸
for score in scores_list:
    # í•˜ë‚˜ì”© êº¼ë‚´ì„œ 5ë¥¼ ë”í•˜ê³  ìƒˆ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    new_scores_list.append(score + 5)

print(f"ì›ë³¸ ì ìˆ˜ ë¦¬ìŠ¤íŠ¸: {scores_list}")
print(f"5ì  ì˜¬ë¦° í›„ ë¦¬ìŠ¤íŠ¸: {new_scores_list}")
print("\në°ì´í„°ê°€ 5ê°œë§Œ ìˆì–´ë„ ì´ë ‡ê²Œ ë²ˆê±°ë¡œìš´ë°, 100ë§Œ ê°œë¼ë©´...?")
````

## âœ¨ ëª¨ë“ˆ 2: NumPy ë°°ì—´(Array)ì˜ ë“±ì¥

- **NumPyë€?**
  * Numerical Pythonì˜ ì¤„ì„ë§
  * ëŒ€ê·œëª¨ ìˆ«ì ë°ì´í„°ë¥¼ ë¹ ë¥´ê³  íš¨ìœ¨ì ìœ¼ë¡œ ë‹¤ë£° ìˆ˜ ìˆê²Œ í•´ì£¼ëŠ” ë¼ì´ë¸ŒëŸ¬ë¦¬
- **NumPy ë°°ì—´ (ndarray)**
  * NumPyì—ì„œ ì‚¬ìš©í•˜ëŠ” ë‹¤ì°¨ì› ë°°ì—´(ë¦¬ìŠ¤íŠ¸)
  * ì œì•½: **ê°™ì€ ì¢…ë¥˜ì˜ ë°ì´í„°ë§Œ ì €ì¥**
    * ìˆ«ìë©´ ìˆ«ì, ë¬¸ìë©´ ë¬¸ì
    * ì´ ì œì•½ì´ ìˆëŠ” ëŒ€ì‹ , ì†ë„ê°€ ë§¤ìš° ë¹ ë¦…ë‹ˆë‹¤!
  - **ë²¡í„°í™” ì—°ì‚° (Vectorized Operation)**
    * ë°°ì—´ì˜ ëª¨ë“  ìš”ì†Œì— ëŒ€í•´ ë°˜ë³µë¬¸ ì—†ì´ í•œ ë²ˆì— ì—°ì‚°ì„ ì ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
    * ê°€ì¥ ì¤‘ìš”í•œ íŠ¹ì§•!

````python
import numpy as np # ê´€ìŠµì ìœ¼ë¡œ npë¼ëŠ” ë³„ëª…ìœ¼ë¡œ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.

# íŒŒì´ì¬ ë¦¬ìŠ¤íŠ¸ë¥¼ NumPy ë°°ì—´ë¡œ ë³€í™˜
scores_array = np.array([80, 95, 72, 100, 88])

print(f"NumPy ë°°ì—´: {scores_array}")
print(f"ìë£Œí˜•: {type(scores_array)}")
````

````python
# NumPy ë°°ì—´ì„ ì´ìš©í•´ ëª¨ë“  ì ìˆ˜ 5ì  ì˜¬ë¦¬ê¸°
# ë°˜ë³µë¬¸ì´ ì „í˜€ í•„ìš” ì—†ìŠµë‹ˆë‹¤!

new_scores_array = scores_array + 5

print(f"ì›ë³¸ ì ìˆ˜ ë°°ì—´: {scores_array}")
print(f"5ì  ì˜¬ë¦° í›„ ë°°ì—´: {new_scores_array}")

# ë§ì…ˆë¿ë§Œ ì•„ë‹ˆë¼ ëª¨ë“  ì‚°ìˆ  ì—°ì‚°ì´ ê°€ëŠ¥í•©ë‹ˆë‹¤.
print(f"\nëª¨ë“  ì ìˆ˜ 10% ì¸ìƒ: {scores_array * 1.1}")
print(f"ëª¨ë“  ì ìˆ˜ 2ë¡œ ë‚˜ëˆ„ê¸°: {scores_array / 2}")
````

## ğŸš€ ëª¨ë“ˆ 3: ë¨¸ì‹ ëŸ¬ë‹ì„ ìœ„í•œ NumPy í•µì‹¬ ê¸°ëŠ¥

### 1. ë°°ì—´ì˜ í˜•íƒœ (Shape)
- **Shapeì´ë€?**
  * ë°°ì—´ì´ ëª‡ ê°œì˜ í–‰ê³¼ ì—´ë¡œ ì´ë£¨ì–´ì ¸ ìˆëŠ”ì§€ ì•Œë ¤ì£¼ëŠ” ì •ë³´ì…ë‹ˆë‹¤.
  * ë¨¸ì‹ ëŸ¬ë‹ì—ì„œ ë°ì´í„°ì˜ êµ¬ì¡°ë¥¼ íŒŒì•…í•˜ëŠ” ë° ê°€ì¥ ê¸°ë³¸ì ì´ê³  ì¤‘ìš”í•©ë‹ˆë‹¤.
- **1ì°¨ì› ë°°ì—´**: ë²¡í„°(Vector)
  * ex) í•™ìƒ 5ëª…ì˜ ì ìˆ˜
- **2ì°¨ì› ë°°ì—´**: í–‰ë ¬(Matrix)
  * ex) 5ëª… í•™ìƒì˜ êµ­ì–´, ì˜ì–´ 2ê³¼ëª© ì ìˆ˜
- **3ì°¨ì› ë°°ì—´**: í…ì„œ(Tensor)
  * ex) (ê°€ë¡œ, ì„¸ë¡œ, ì»¬ëŸ¬) í”½ì…€ë¡œ ì´ë£¨ì–´ì§„ ì»¬ëŸ¬ ì´ë¯¸ì§€

````python
# 1ì°¨ì› ë°°ì—´ (ë²¡í„°)
# [ìƒì1, ìƒì2, ìƒì3, ìƒì4, ìƒì5, ìƒì6] -> í•œ ì¤„ë¡œ 6ê°œ
array_1d = np.array([1, 2, 3, 4, 5, 6])
print(f"1D ë°°ì—´: {array_1d}")
print(f"1D ë°°ì—´ì˜ Shape: {array_1d.shape}") # (6,) -> ì•„ì´í…œì´ 6ê°œì¸ í•œ ì¤„

print("-" * 20)

# 2ì°¨ì› ë°°ì—´ (í–‰ë ¬)
# [[êµ­ì–´1, ì˜ì–´1],  -> ì²« ë²ˆì§¸ ì¤„ (í–‰)
#  [êµ­ì–´2, ì˜ì–´2],  -> ë‘ ë²ˆì§¸ ì¤„ (í–‰)
#  [êµ­ì–´3, ì˜ì–´3]] -> ì„¸ ë²ˆì§¸ ì¤„ (í–‰)
array_2d = np.array([
    [90, 85], # í•™ìƒ 1
    [75, 92], # í•™ìƒ 2
    [88, 80]  # í•™ìƒ 3
])
print(f"2D ë°°ì—´:\n{array_2d}")
print(f"2D ë°°ì—´ì˜ Shape: {array_2d.shape}")
# (3, 2) -> 3ê°œì˜ í–‰, 2ê°œì˜ ì—´
````

### 2. ë°°ì—´ì˜ í˜•íƒœ ë°”ê¾¸ê¸° (Reshape)
- `reshape()`
  * ë°ì´í„°ì˜ ë‚´ìš©ê³¼ ì´ ê°œìˆ˜ëŠ” ê·¸ëŒ€ë¡œ ë‘” ì±„,
  * ë°°ì—´ì˜ í˜•íƒœ(ê»ë°ê¸°)ë§Œ ë°”ê¿€ ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤.
- **-1ì˜ í™œìš©**
  * **"ë‚˜ë¨¸ì§€ëŠ” ë„¤ê°€ ì•Œì•„ì„œ ê³„ì‚°í•´ì¤˜"** ë¼ëŠ” ì˜ë¯¸ì˜ ì™€ì¼ë“œì¹´ë“œì…ë‹ˆë‹¤.
  - `array.reshape(2, -1)`
    * "ì´ 6ê°œì¸ë°, 2ê°œì˜ í–‰ìœ¼ë¡œ ë§Œë“¤ê³  ì‹¶ì–´. ì—´ì€ ì•Œì•„ì„œ ê³„ì‚°í•´ì¤˜."
    * â†’ (2, 3)ìœ¼ë¡œ ë³€í™˜
  - `array.reshape(-1, 3)`
    * "ì´ 6ê°œì¸ë°, 3ê°œì˜ ì—´ë¡œ ë§Œë“¤ê³  ì‹¶ì–´. í–‰ì€ ì•Œì•„ì„œ ê³„ì‚°í•´ì¤˜."
    * â†’ (2, 3)ìœ¼ë¡œ ë³€í™˜
  - **ì£¼ì˜**: ì „ì²´ ì›ì†Œ ê°œìˆ˜ê°€ ë‚˜ëˆ„ì–´ ë–¨ì–´ì§€ì§€ ì•Šìœ¼ë©´ ì—ëŸ¬ê°€ ë°œìƒí•©ë‹ˆë‹¤.
    * ex) 6ê°œ ì›ì†Œë¥¼ 4ê°œì˜ í–‰ìœ¼ë¡œ ë§Œë“¤ ìˆ˜ ì—†ìŒ

````python
# 1ì°¨ì› ë°°ì—´ (ì•„ì´í…œ 6ê°œ)
array_1d = np.array([1, 2, 3, 4, 5, 6])
print(f"ì›ë³¸ 1D ë°°ì—´: {array_1d}")
print(f"ì›ë³¸ 1D ë°°ì—´ Shape: {array_1d.shape}")

# 2í–‰ 3ì—´ì˜ 2ì°¨ì› ë°°ì—´ë¡œ í˜•íƒœ ë³€ê²½
# [1,2,3,4,5,6] -> [[1,2,3],
#                   [4,5,6]]
array_reshaped = array_1d.reshape(2, 3)
print(f"\n(2, 3)ìœ¼ë¡œ Reshape:\n{array_reshaped}")
print(f"Shape: {array_reshaped.shape}")

# -1ì„ ì‚¬ìš©í•˜ì—¬ ì—´ì˜ ê°œìˆ˜ë¥¼ ìë™ìœ¼ë¡œ ê³„ì‚°
# "6ê°œ ì›ì†Œë¥¼ 3ê°œì˜ í–‰ìœ¼ë¡œ ë§Œë“¤ì–´ì¤˜. ì—´ì€ ì•Œì•„ì„œ!" -> 2ê°œì˜ ì—´ í•„ìš”
array_reshaped_auto = array_1d.reshape(3, -1)
print(f"\n(3, -1)ìœ¼ë¡œ Reshape:\n{array_reshaped_auto}")
print(f"Shape: {array_reshaped_auto.shape}")
````

### 3. ì¶•(Axis)ì„ ë”°ë¥¸ ì—°ì‚°
- **ì¶•(Axis)ì´ë€?**
  * ë‹¤ì°¨ì› ë°°ì—´ì—ì„œ **ì—°ì‚°ì„ ìˆ˜í–‰í•  ë°©í–¥**ì„ ì§€ì •í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤.
- **ë¨¸ë¦¿ì†ìœ¼ë¡œ ê·¸ë¦¬ê¸° (2ì°¨ì› ê¸°ì¤€)**
  - **`axis=0`**: **ì„¸ë¡œ ë°©í–¥(â†“)**
    * ê° **ì—´(column)**ì— ëŒ€í•´ ì—°ì‚°í•©ë‹ˆë‹¤.
    * ì—¬ëŸ¬ í–‰ë“¤ì„ í•˜ë‚˜ì˜ í–‰ìœ¼ë¡œ ê¾¹ ëˆŒëŸ¬ ì••ì¶•í•˜ëŠ” ëª¨ìŠµì„ ìƒìƒí•˜ì„¸ìš”.
  - **`axis=1`**: **ê°€ë¡œ ë°©í–¥(â†’)**
    * ê° **í–‰(row)**ì— ëŒ€í•´ ì—°ì‚°í•©ë‹ˆë‹¤.
    * ì—¬ëŸ¬ ì—´ë“¤ì„ í•˜ë‚˜ì˜ ì—´ë¡œ ê¾¹ ëˆŒëŸ¬ ì••ì¶•í•˜ëŠ” ëª¨ìŠµì„ ìƒìƒí•˜ì„¸ìš”.
- `axis`ë¥¼ ì§€ì •í•˜ì§€ ì•Šìœ¼ë©´, ë°°ì—´ì˜ ëª¨ë“  ì›ì†Œì— ëŒ€í•´ ì—°ì‚°í•©ë‹ˆë‹¤.

````python
# 3ëª… í•™ìƒì˜ êµ­ì–´, ì˜ì–´ ì ìˆ˜
scores_2d = np.array([
    [90, 85], # í•™ìƒ 1
    [75, 92], # í•™ìƒ 2
    [88, 80]  # í•™ìƒ 3
])
print(f"ì›ë³¸ ì ìˆ˜ ë°°ì—´ (Shape: {scores_2d.shape}):\n{scores_2d}")

# 1. ì „ì²´ í•©ê³„ / í‰ê· 
print(f"\nì „ì²´ ì´ì : {scores_2d.sum()}")
print(f"ì „ì²´ í‰ê· : {scores_2d.mean():.2f}")

# 2. axis=0 (ì„¸ë¡œ ë°©í–¥ â†“) ì—°ì‚°
# ê° 'ì—´'ì˜ í•©ê³„/í‰ê·  -> ê³¼ëª©ë³„ í†µê³„
print(f"\n--- axis=0 (ê³¼ëª©ë³„ í†µê³„) ---")

print(f"ê³¼ëª©ë³„ ì´ì : {scores_2d.sum(axis=0)}")
# [90+75+88, 85+92+80]

print(f"ê³¼ëª©ë³„ í‰ê· : {scores_2d.mean(axis=0)}")
# [avg(90,75,88), avg(85,92,80)]

print(f"ê³¼ëª©ë³„ ìµœê³ ì : {scores_2d.max(axis=0)}")
# [max(90,75,88), max(85,92,80)]

# 3. axis=1 (ê°€ë¡œ ë°©í–¥ â†’) ì—°ì‚°
# ê° 'í–‰'ì˜ í•©ê³„/í‰ê·  -> í•™ìƒë³„ í†µê³„
print(f"\n--- axis=1 (í•™ìƒë³„ í†µê³„) ---")

print(f"í•™ìƒë³„ ì´ì : {scores_2d.sum(axis=1)}")
# [90+85, 75+92, 88+80]

print(f"í•™ìƒë³„ í‰ê· : {scores_2d.mean(axis=1)}")
# [avg(90,85), avg(75,92), avg(88,80)]

print(f"í•™ìƒë³„ ìµœê³ ì : {scores_2d.max(axis=1)}")
# [max(90,85), max(75,92), max(88,80)]
````
