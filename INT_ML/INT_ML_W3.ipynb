### 1. ë¦¬ìŠ¤íŠ¸ (List)
- **ì •ì˜**:

- **ì˜ˆì‹œ**: ëª¨ë“  í•™ìƒì˜ 'ê³µë¶€ ì‹œê°„' / 'ì‹œí—˜ì ìˆ˜' ë¦¬ìŠ¤íŠ¸

````python
# ì—¬ëŸ¬ í•™ìƒì˜ 'ê³µë¶€ ì‹œê°„'ê³¼ 'ì‹œí—˜ ì ìˆ˜' ë°ì´í„°ë¥¼ ë¦¬ìŠ¤íŠ¸ë¡œ ë§Œë“¤ì–´ ë´…ì‹œë‹¤.
study_hours_list = [2.0, 4.5, 6.0, 8.5, 10.0]
exam_scores_list = [65, 78, 85, 95, 98]

print("ê³µë¶€ ì‹œê°„ ë¦¬ìŠ¤íŠ¸:", study_hours_list)
print("ì‹œí—˜ ì ìˆ˜ ë¦¬ìŠ¤íŠ¸:", exam_scores_list)

# ë¦¬ìŠ¤íŠ¸ëŠ” 'ìˆœì„œ'ê°€ ìˆê¸° ë•Œë¬¸ì—, ì¸ë±ìŠ¤(index)ë¡œ íŠ¹ì • ê°’ì— ì ‘ê·¼í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
# íŒŒì´ì¬ì€ 0ë¶€í„° ìˆ«ìë¥¼ ì…‰ë‹ˆë‹¤!
print("\nì²« ë²ˆì§¸ í•™ìƒì˜ ê³µë¶€ ì‹œê°„:", study_hours_list[0])
print("ì„¸ ë²ˆì§¸ í•™ìƒì˜ ì‹œí—˜ ì ìˆ˜:", exam_scores_list[2])

# len() í•¨ìˆ˜ë¡œ ë¦¬ìŠ¤íŠ¸ì— ëª‡ ê°œì˜ ë°ì´í„°ê°€ ìˆëŠ”ì§€ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
print("\nì´ í•™ìƒ ìˆ˜:", len(study_hours_list))
````

````python
# ìƒˆë¡œìš´ í•™ìƒì˜ ë°ì´í„°ê°€ ì¶”ê°€ë˜ì—ˆë‹¤ê³  ê°€ì •í•´ë´…ì‹œë‹¤.
# .append()ë¥¼ ì´ìš©í•´ ë¦¬ìŠ¤íŠ¸ì˜ ë§¨ ë’¤ì— ê°’ì„ ì¶”ê°€í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

print("ì¶”ê°€ ì „ ê³µë¶€ ì‹œê°„ ë¦¬ìŠ¤íŠ¸:", study_hours_list)

study_hours_list.append(12.0)
exam_scores_list.append(100)

print("ì¶”ê°€ í›„ ê³µë¶€ ì‹œê°„ ë¦¬ìŠ¤íŠ¸:", study_hours_list)
print("ì¶”ê°€ í›„ ì´ í•™ìƒ ìˆ˜:", len(study_hours_list))
````

### 2. ë”•ì…”ë„ˆë¦¬ (Dictionary)

  - 'í‚¤(Key)'ì™€ 'ê°’(Value)'ì„ í•˜ë‚˜ì˜ ìŒìœ¼ë¡œ ë¬¶ì–´ì„œ ì €ì¥í•˜ëŠ” ìë£Œêµ¬ì¡°

- **ì˜ˆì‹œ**: í•œ í•™ìƒì˜ 'ì´ë¦„', 'ê³µë¶€ ì‹œê°„', 'ì‹œí—˜ ì ìˆ˜', 'í•©ê²© ì—¬ë¶€'ë¥¼ ëª¨ë‘ ë‹´ê³  ìˆëŠ” ì •ë³´ ë¬¶ìŒ

````python
# í•œ í•™ìƒì˜ ëª¨ë“  ì •ë³´ë¥¼ ë”•ì…”ë„ˆë¦¬ë¡œ ë§Œë“¤ì–´ ë´…ì‹œë‹¤.
student_A_info = {
    'name': 'Park',
    'study_hours': 8.5,
    'exam_score': 95,
    'is_passed': True
}

print("í•™ìƒ Aì˜ ì •ë³´:", student_A_info)

# ë”•ì…”ë„ˆë¦¬ëŠ” ìˆœì„œ(index) ëŒ€ì‹  'í‚¤(Key)'ë¥¼ ì´ìš©í•´ ê°’ì— ì ‘ê·¼í•©ë‹ˆë‹¤.
print("\ní•™ìƒ Aì˜ ì´ë¦„:", student_A_info['name'])
print("í•™ìƒ Aì˜ ì‹œí—˜ ì ìˆ˜:", student_A_info['exam_score'])
````
